## Star Wars Json Challenge

Sua tarefa é criar um script que abra esse arquivo, extraia seu conteúdo, exiba o nome e o gênero de cada personagem e, ao final, mostre a quantidade de personagens do gênero masculino, a quantidade de personagens do gênero feminino e a quantidade de personagens de gênero não indicado presentes nesse arquivo.


Para realizar essa tarefa, você deve utilizar apenas as estruturas de programação na linguagem Python versão 3 que foram estudadas até esse ponto. Encare essa limitação como um desafio para exercitar suas habilidades de programação aliadas à sua criatividade.

```python
import json

file_path = '/home/borges/dev/level2/files/challenge/star_wars.json'

try:
    with open(file_path, 'r') as file:
        read = file.read()
finally:
    file.close()

contents = json.loads(read)

male_count = 0
female_count = 0
no_gender_count = 0

for obj in contents:
    print(f'{obj['name']} is {obj['gender']} gender')
    if(obj['gender'] == 'male'):
        male_count += 1
    elif(obj['gender'] == 'female'):
        female_count += 1
    else:
        no_gender_count += 1

print(f'\n\nThere are {female_count} female characters, {male_count} male characters and {no_gender_count} characters with no gender assigned. \nTotal: {female_count + male_count+ no_gender_count} out of {len(contents)} characters count')
```
